```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
branch_visits = pd.read_csv("../data/cleaned_files/tpl_branch_visits_2012_2022_clean.csv")
branch_visits.info()     
```

```python
branch_reg = pd.read_csv("../data/cleaned_files/tpl_branch_registrations_2012_2022_clean.csv")
branch_reg.info()    
```

```python
branch_circ = pd.read_csv("../data/cleaned_files/tpl_branch_circulation_2012_2022_clean.csv")
branch_circ.info()    
```

```python
branch_wrkstn_use = pd.read_csv("../data/cleaned_files/tpl_branch_workstation_use_2012_2022_clean.csv")
branch_wrkstn_use.info()   
```

```python
branch_lvl_stats_1 = pd.merge(branch_visits, branch_reg, on=['NBHDNo', 'NBHDName', 'BranchCode', 'BranchName','Year'], how = 'outer')
branch_lvl_stats_1
```

```python
branch_lvl_stats_2 = pd.merge(branch_lvl_stats_1, branch_circ, on=['NBHDNo', 'NBHDName', 'BranchCode', 'BranchName','Year'], how = 'outer')
branch_lvl_stats_2
```

```python
branch_lvl_stats_3 = pd.merge(branch_lvl_stats_2, branch_wrkstn_use, on=['NBHDNo', 'NBHDName', 'BranchCode', 'BranchName','Year'], how = 'outer')
branch_lvl_stats_3
```

```python
# Convert 'NBHDNo' to string and add leading zeros
branch_lvl_stats_3['NBHDNo'] = branch_lvl_stats_3['NBHDNo'].apply(lambda x: str(x).zfill(3))
branch_lvl_stats_3
```

```python
branch_lvl_stats_3 = branch_lvl_stats_3.fillna(0)
```

```python
branch_lvl_stats = branch_lvl_stats_3.copy()
```

```python
branch_lvl_stats.to_csv('../data/cleaned_files/branch_lvl_stats.csv', index=False)
```
